Task 1. 2: Similarly to the “while” loop in section 3, write a “while” loop based on the value
of randomly generated number being higher than 0.8, add following functionality:
1) After the loop ends, print how many iterations it went through
2) Based on naive expected number of iterations (e.g. for 0.5 we “expect” 1 pass for every 2
numbers, so needing more than 2 would be “unlucky”), after the loop ends print “lucky” or
“unlucky” message.

```python
import random
i = 0
while random.random() < 0.9:
  i += 1
if i > 1 / (1 - 0.9):
  print('unlucky')
else:
  print('lucky')
print(f"it took {i} iterations")
```

Task 1.4: Write a program that solves the following equation (the solution can be
approximate):
3x2 – 4x - 2 = 0.

```python
from math import sqrt

a = 3
b = -4
c = -2

d = b**2 - 4 *a*c

x1 = (-b - sqrt(d)) / (2 * a)
x2 = (-b + sqrt(d)) / (2 * a)

print(x1)
print(x2)
```

Task 1.5: Write a function for solving quadratic equations, use it to solve task 1.4

```python
def quad_solver(a,b,c):
  delta = b**2 - 4*a*c
  if(delta > 0):
    x1 = (-b - sqrt(delta)) / (2 * a)
    x2 = (-b + sqrt(delta)) / (2 * a)

    print(f"X1 = {x1}")
    print(f"X2 = {x2}")
  elif(delta == 0):
    x = -b / (2 * a)
    print(x)
  else:
    print("No real roots")

quad_solver(3,-4,-2)
quad_solver(1,3,1)
quad_solver(3,1,10)
```

Write code that separates the values in both values1 and values2 arrays into two lists (for
information on how to use lists, refer to Example code in section Longer code examples,
both “circumferences” and “areas” are lists), one containing values lower than 0.5, the other

```python
import numpy as np
values1 = np.random.rand(20)
values2 = np.random.rand(20)
```

```python
lower = [] #than 0.5
higher = []

for i in range(len(values1)):
  if(values1[i] < 0.5):
    lower.append(values1[i])
  else:
    higher.append(values1[i])
  if(values2[i] < 0.5):
    lower.append(values2[i])
  else:
    higher.append(values2[i])

print("Lower list:")
print(lower)
print(f"Lower count: {len(lower)}")
print(higher)
print(f"Higher count: {len(higher)}")
```
